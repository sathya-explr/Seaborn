#### Distribution plots

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
sales = pd.read_excel('sales.xlsx')
```

```python
sales.head(4)
```

```python
# first distribution plot
sns.distplot(sales.Profit)
```

```python
sns.distplot(sales.Profit,kde=False)
# to remove the outlinr
```

kde= kernel density estimation. KDE= False removes the outline  and hist=False removes the shaded region

```python
sns.distplot(sales.Profit,hist=False)
```

```python
# rug parameter, shows the concentration of data in a distribution plot
sns.distplot(sales.Profit,kde=False,rug=True,bins=10)
```

```python
# in kde plot you can give parameter bw (bandwidth)
sns.kdeplot(sales.Profit,bw = 10)
```

##### Didtribution using scipy

```python
from scipy import stats
```

```python
# to plot gamma distribution, we can get the vairiable from numpy
a = np.random.gamma(6,size=200)
```

```python
sns.distplot(a,fit=stats.gamma)
```

Blue line is the kde, you can remove it by giving kde=False

#### Joint Plot
> It gives the distribution of both x and y axis along with scatter plot

```python
sns.jointplot('Sales','Discount',data=sales)
```

```python
sns.jointplot('Discount','Product Base Margin',data=sales,kind='hex')
```
