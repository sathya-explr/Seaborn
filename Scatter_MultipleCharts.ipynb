```python
import pandas as pd
import seaborn as sns
```

```python
sales = pd.read_excel('sales.xlsx')
sales.head(5)
```

```python
# Basic scatter plot
sns.relplot('Sales','Profit',data=sales)
```

#### Using col parameter
It is for getting seperate chart for each of the categorical values of the said column

```python
# here we use col for Customer Segment
sns.relplot('Sales','Profit',data=sales, col='Customer Segment')
```

```python
sales['Customer Segment'].unique()
```

```python
# use col_wrap for getting clear images
sns.relplot('Sales','Profit',data=sales,kind='line', col='Customer Segment',col_wrap = 2)
```

#### Using row parameter
here we use row along with col, the graphs formes are like table, where each represents a particule row column value combination
hue is for color differentiation based on 

```python
sns.relplot('Sales','Profit',data=sales, col='Customer Segment', row = 'Product Category',hue = 'Ship Mode')
```
