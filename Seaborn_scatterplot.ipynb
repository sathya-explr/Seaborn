```python
import pandas as pd
import seaborn as sns
```

```python
sales = pd.read_excel('sales.xlsx')
```

```python
sales.head(5)
```

#### Relationship plot or Scatter plot
It requires three parameters, x val, y val, and the source dataset

sns.relplot('Sales','Profit',data = sales)

#### To  know strength based on a variable. colorcoding

```python
# use parameter hue
sns.relplot('Sales','Profit',data=sales, hue='Order Priority')

# here there is no need to give legend
```

```python
# to differentiate the priority categoried based on icon marks use style parameter
sns.relplot('Sales','Profit',data=sales, hue='Order Priority', style='Order Priority')
```

```python
# It is possible to use different columns or parameters for color coding and style
sns.relplot('Sales','Profit',data=sales, hue='Order Priority', style='Ship Mode')
```

```python
# The size parameter allows you to vary the sizes based on the data value
sns.relplot('Sales','Profit',data=sales,size = 'Discount')
```

```python
# tuning of sizes For more clarity, use the attribute sizes
sns.relplot('Sales','Profit',data=sales,size = 'Discount', sizes =(25,200))
```
