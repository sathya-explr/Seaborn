```python
import pandas as pd
import seaborn as sns
```

```python
sales = pd.read_excel('sales.xlsx')
```

```python
sns.catplot('Product Category','Shipping Cost',data=sales,kind='violin')
```

```python
# inner = 'stick' to  show the density of data, but use it when the data size is small, no. ofrows is less.
sns.catplot('Product Category','Shipping Cost', hue = 'Ship Mode',inner= 'stick',data=sales,kind='violin')
```

In the above plot each line in the violin represents a row.

#### Bar Chart

```python
sns.catplot(x='Product Category', y='Sales',data=sales,kind='bar')
```

```python
sns.catplot(x='Product Category', y='Sales',data=sales,hue ='Ship Mode',kind='bar')
```

#### Point Plot

```python
sns.catplot(x='Product Category', y='Sales',data=sales,hue='Ship Mode',kind='point')
# the line on the point shows the deviation
```

```python
sns.catplot(x='Product Category', y='Sales',col='Region',col_wrap= 3,data=sales, kind='point')
```

```python
sns.catplot(x='Product Category', y='Sales',data=sales,hue ='Ship Mode',col='Customer Segment',kind='point')
```
