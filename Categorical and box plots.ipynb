```python
import pandas as pd
import seaborn as sns
```

```python
sales = pd.read_excel('sales.xlsx')
```

```python
sales.head(4)
```

```python
sns.catplot('Ship Mode','Sales',data=sales)
```

In the above plot, the density of dots implies more sales happening

```python
# to remove graph width use jitter argument
sns.catplot('Ship Mode','Sales',data=sales, jitter = False)
```

```python
# including conditionals in the plot statement df_name.query() fn
sns.relplot('Ship Mode','Sales',data=sales.query('Province == "Nunavut"'))
```

```python
sns.catplot('Ship Mode','Sales',data=sales.query('Profit> 1000'), jitter = False)
```

#### Swarm Plots:
These show magnitude

```python
sns.catplot('Ship Mode','Sales',data=sales.query('Profit> 1000'), kind = 'swarm')
```

#### Box Plot

```python
sns.catplot(x = 'Product Category',y = 'Shipping Cost', data=sales, kind = 'box' )
```

```python
sns.catplot(x = 'Product Category',y = 'Shipping Cost', hue = 'Ship Mode', data=sales, kind = 'box' )
```

```python
# ordering in the hue , (format)
sns.catplot(x = 'Product Category',y = 'Shipping Cost', hue = 'Ship Mode', 
            hue_order=['Delivery Truck','Regular Air','Express Air'],
            data=sales, kind = 'box')
```

```python
sns.catplot(x = 'Product Category',y = 'Shipping Cost', col = 'Ship Mode', data=sales, kind = 'box' )
# for each ship mode a separate chart
```
